```python
import pandas as pd
```

```python
dataset=pd.read_csv('insurance_pre.csv')
```

```python
#Retriving columns present in the Dataset
dataset
```

```python
#Performing One Hot Encoding 
dataset= pd.get_dummies(dataset)
```

```python
dataset.columns
```

```python
#Splitting the input columns from the dataset
independent=dataset[['age', 'bmi', 'children','sex_female', 'sex_male','smoker_no', 'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(n_estimators=2000,criterion='friedman_mse',max_features='log2',random_state=None)
regressor.fit(x_train,y_train)
```

```python
y_predict=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score=r2_score(y_test,y_predict)
r2_score
```
